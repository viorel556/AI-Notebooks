#### Simple Neural Networks in TensorFlow

Using TensorFlow to create a ultra simple 2 layer neural network and doing inference over it:
 - first layer has 3 units (neurons)
 - second layer has just one unit 
 - layer 1 receives a 2-dimensional vector (with 2 values only)

```python
from tensorflow.keras.layers import Dense 
from tensorflow.keras import Sequential  
from viorel_utils import see 
import numpy as np 
```

#### Example (1)

##### Scheme: 
 


```python
# getting the input vector data also, also called a[0]
x = np.array([[200, 17]])
```

```python
# creating layer 1:  
layer_1 = Dense(units = 3, activation = 'sigmoid')

# computing a1: (vector resulting from layer 1)
a1 = layer_1(x) 

# creating layer 2: 
layer_2 = Dense(units = 1, activation = 'sigmoid')

# computing a2 (by getting last vector)
a2 = layer_2(a1)
```

##### Results:
Will be shown conditionally based on our set treshold. In this case is 0.5   

```python
# predicted y 
yhat = any   

if a2 > 0.5: 
  yhat = 1 
else: 
  yhat = 0 

# [^] Can be also written as ternary expression: 
yhat = 1 if a2 > 0.5 else 0 

see(yhat)
```

#### Example (2)

##### Scheme: 


- 3 layer neural network; 
- first layer has 25 neurons (units) 
- second layer has 15 units 
- third layer has only one unit 
- receives a vector x, aka a[0]

```python
# DA NEURAL NETWORK: 
x = np.array([[0.0, 245, 240, 0, 200, 10, 22, 10]])
# | 
layer_1 = Dense(units=25,activation='sigmoid')
# | 
a1 = layer_1(x) # a1 is a matrix 
# | 
layer_2 = Dense(units=15, activation='sigmoid')
# | 
a2 = layer_2(a1)
# |
layer_3 = Dense(units=1, activation='sigmoid')
# | 
a3 = layer_3(a2)
# | 
yhat = any 

# Results: 
yhat = 1 if a3 >= 0.5 else 0 

see(yhat)
see(a2)
```
